```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3

%matplotlib inline
```

```python
conn = sqlite3.connect(':memory:')
df = pd.read_csv('ai_job_dataset.csv')
df.to_sql("AI", conn, index = False, if_exists = 'replace')
```

```python
query = ('''SELECT * FROM AI''')
result = pd.read_sql_query(query, conn)
result
```

```python
print(df.columns.to_list())
```

```python
ax = df['salary_usd'].plot(kind = 'density', figsize = (10,3))
ax.axvline(df['salary_usd'].mean(), color = 'red')
ax.axvline(df['salary_usd'].median(), color = 'green')
```

```python
df['salary_usd'].plot(kind = 'hist', figsize = (10,3))
```

```python
df.plot(kind = 'scatter', x = 'job_title', y = 'salary_usd', figsize = (10,3))
```

```python
ax = df['job_title'].value_counts().plot(kind = 'bar', figsize = (10,3))
ax.set_ylabel('salary_usd')
```
